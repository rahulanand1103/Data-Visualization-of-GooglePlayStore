### Attribute information
<ul>
<li><b>App</b>- Application name</li>
<li><b>Category</b>- Category the app belongs to
<li><b>Rating</b>-Overalluser rating of the app (as when scraped)</li>
<li><b>Reviews</b>- Number-of user reviews for the app (as when scraped)</li>
<li><b>Size</b>- Size of the app (as when scraped)</li>
<li><b>Installs</b>-Number of user downloads/installs for the app (as when scraped)</li>
<li><b>Type</b>- Paid or Free</li>
<li><b>Price</b>- Price of the app (as when scraped)</li>
<li><b>Content Rating</b>- Age group the app is targeted at - Children / Mature 21+ / Adult</li>
<li><b>Genres</b>-An app can belong to multiple genres (apart from its main category). For eg, a musical family game will belong to Music, Game, Family genres.</li>
<li><b>Last Updated</b>-Date when the app was last updated on Play Store (as when scraped)</li>
<li><b>Current Vers</b>-Current version of the app available on Play Store (as when scraped)</li>
<li><b>Android Ver</b>-Min required Android version (as when scraped)</li>

## 3.1 Data Loading

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#ignore warning messages
import warnings
warnings.filterwarnings('ignore')
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.graph_objs as go
from wordcloud import WordCloud
```

```python
df=pd.read_csv("googleplaystore.csv")
df
```

## 3.2 Data Preprocessing

```python
#total no of Apps before cleaning
print(df.shape)
#Checking null value
print(df.isnull().values.any())
```

```python
#dropping the rows which has null value 
df = df.dropna(how='any',axis=0) 
```

```python
#printing shape after removing the rows which has null values
print(df.shape)
```

```python
#removing the dupilcate apps
df.drop_duplicates(subset='App', inplace=True)
```

### After Preprocessing Number of rows

```python
print(df.shape)
```

### App Installs more than 500,000,000+,1,000,000,000+,100,000,000+

```python
Apps1000K=df[df['Installs'] =="1,000,000,000+"]
Apps500K=df[df['Installs'] == "500,000,000+"]
Apps100K=df[df['Installs'] == "100,000,000+"]
fames=[Apps1000K, Apps500K, Apps100K]
topApp = pd.concat(fames)
```

### 5. App Installs more than 1,000,000,000+

```python
MostDownloadedApp=[]
for x in Apps1000K['App']:
    MostDownloadedApp.append(x.replace(" ",""))
wordcloud = WordCloud(width = 800, height = 800,
            background_color ='white').generate(str(MostDownloadedApp))
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```

```python
Category=Apps1000K['Category'].value_counts()
plt.figure(figsize=(12,12))
g=sns.barplot(Category.index, Category.values, alpha=0.8)
g.set_xticklabels(rotation=90,labels=Category.index)
plt.title('Content Rating vs No Apps')
plt.ylabel('Number of Apps')
plt.xlabel('Category')
plt.show()
```
